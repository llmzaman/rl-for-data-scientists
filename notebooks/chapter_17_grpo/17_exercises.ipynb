### Exercise 17.1: Group size G sensitivityFor GRPO, plot the variance of advantage estimates as a function of group size G ∈ {2, 4, 8, 16, 32}. At what G does variance stabilise? How does this depend on task difficulty (fraction of correct responses)?

```python
import numpy as np, matplotlib.pyplot as plt

def advantage_variance_by_G(p_correct, G_values, n_trials=1000):
    """Simulate advantage variance as a function of group size."""
    variances = []
    for G in G_values:
        trial_vars = []
        for _ in range(n_trials):
            rewards = np.random.binomial(1, p_correct, G).astype(float)
            if rewards.std() < 1e-8:
                trial_vars.append(0)
            else:
                adv = (rewards - rewards.mean()) / rewards.std()
                trial_vars.append(adv.var())
        variances.append(np.mean(trial_vars))
    return variances

# TODO: plot for p_correct ∈ {0.1, 0.3, 0.5, 0.7, 0.9}
```
